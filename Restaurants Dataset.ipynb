```python
# Using pandas, sqlalchemy and numpy in this analysis
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
# Importing csv files to create dataframe
rest_file = "../data/dataset.csv"
data_df = pd.read_csv(rest_file)

# removing null values to avoid errors  
data_df.dropna(how ='all', inplace = True) 
data_df
```

```python
data_df["CAMIS"].drop_duplicates(keep='last')

data_df
```

```python
#data_df = pd.dataFrame(data_df)
# data_df.nunique(axis=1)
# data_df.head()
```

```python
# using str.lower() to change letter case 

new= data_df.loc[:,['CAMIS','DBA','STREET']]
new['DBA']= new['DBA'].str.lower()
new['STREET']= new['STREET'].str.lower()
#dba.to_append()
#dba.join(nta)
data= pd.merge(data_df, new, on='CAMIS', how= 'outer')
data.head()
```

```python
# Cleaning dataset by dropping useless columns and combine all same ids
newdata= data.drop(['DBA_x','STREET_x','GRADE DATE','RECORD DATE','Community Board','Council District', 'BIN','BBL','NTA'], axis=1)

newdata.head()
```

```python
# Rename columns to lowercase for sql
newdata = newdata.rename(columns={"CAMIS": "record_id", "DBA_y" : "name", "BORO":"borough", "BUILDING":"building", "STREET_y":"street",
                                  "ZIPCODE":"zipcode","PHONE":"phone","CUISINE DESCRIPTION":"cuisine",
                                  "INSPECTION DATE":"inspection_date", "ACTION": "action", "VIOLATION CODE": "violation_code",
                                  "VIOLATION DESCRIPTION":"description",
                                  "CRITICAL FLAG":"flag", "SCORE":"score","GRADE":"grade","INSPECTION TYPE":"inspection_type",
                                 "Latitude":"lat","Longitude":"lng","Census Tract":"census_tract"})
newdata
```

```python
newdata.to_csv('dataset1.csv',index=False)
```
